```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.common.by import By
import time
import os
import urllib.request as req
import xlsxwriter # 엑셀 저장
```

```python
browser= webdriver.Chrome('./chromedriver_win32/chromedriver.exe')
browser.implicitly_wait(5)
browser.get('http://www.yes24.com/main/default.aspx?ysmchn=ggl&ysmcpm=google-sponsor&ysmtac=ppc&ysmtrm=%EC%98%88%EC%8A%A424&pid=123487&cosemkid=go14913756274107498&gclid=Cj0KCQiAuefvBRDXARIsAFEOQ9EGr3URUoNtoshIsLrVxoApamBLsLC_AwBY56b7WjLrVwHvCo4K-uUaAnukEALw_wcB')
```

```python
WebDriverWait(browser, 3).until(ec.presence_of_element_located((By.XPATH, '//*[@id="query"]')))
search = browser.find_element_by_xpath('//*[@id="query"]')
search.send_keys('C#')
browser.find_element_by_xpath('//*[@id="yesSForm"]/fieldset/span[2]/button').click()
```

```python
soup = BeautifulSoup(browser.page_source, 'html.parser')
prodlist = soup.select('div.main_prodlist.main_prodlist_list > ul > li.prod_item.prod_layer')

savePath = './TestTmpFolder/BookImage/'
if not (os.path.isdir(savePath)):
    os.makedirs(os.path.join(savePath))

# 엑셀 처리
workbook = xlsxwriter.Workbook('./TestTmpFolder/yes24_crawling_result.xlsx')
worksheet = workbook.add_worksheet()
```

```python
cur_page_num = 1
target_page_num = 10

row_num=1
while cur_page_num <= target_page_num:
    
    soup = BeautifulSoup(browser.page_source, 'html.parser')
    prodlist = soup.select('div.goodsList.goodsList_list:nth-of-type(2) > table > tbody > tr')
    
    for item in prodlist:
         if not (item.find('td', class_='goods_division')):
            if not (item.select('td.goods_infogrp > p')[0].text.split(']')[0].split('[')[1] == 'eBook') :
                goodscode = item.select('td.goods_img > a')[0].get('href').split('Goods/')[1].split('?')[0]
                goodsname = item.select('td.goods_infogrp > p > a')[0].text
                goodspublisher = item.select('td.goods_infogrp > div.goods_info > a')[-1].text
                goodsdate = item.select('td.goods_infogrp > div.goods_info > em')[0].text
                goodsprice = item.select('td.goods_infogrp > p.goods_price > strong')[0].text
                goodsimg = item.select('td.goods_img > a > img')[0].get('src').split('/95x0')[0]
                print(goodscode,'\n',goodsname,'\n',goodsprice, '\n',goodspublisher,'\n', goodsimg)

                worksheet.write(f'A{row_num}', goodscode)
                worksheet.write(f'B{row_num}', goodsname)
                worksheet.write(f'C{row_num}', goodspublisher)
                worksheet.write(f'D{row_num}', goodsdate)
                worksheet.write(f'E{row_num}', goodsprice)
                row_num +=1

                FileName = os.path.join(savePath +goodscode+'.jpg')
                req.urlretrieve(goodsimg, FileName)
    print()
    cur_page_num += 1
    if(cur_page_num > target_page_num):
        print('크로울링 끝')
        break;
    # 페이지 이동버튼 클릭
    WebDriverWait(browser, 5).until(ec.presence_of_element_located((By.CSS_SELECTOR, f'div.pagen.pat30.pab15 > a:nth-child({ 2 + cur_page_num})'))).click()
    
    del soup
    time.sleep(5)
```
