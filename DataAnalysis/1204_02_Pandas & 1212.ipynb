### 원하는 로우 선택

### 시각화

```python
import matplotlib.pyplot as plt
# 매직 커멘드 : 해당 파일에 그래프를 그려라 (inline)
%matplotlib inline 

# 마이너스 기호가 깨지는 현상 방지
from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

# 한글 폰트 지원하게하기
f_path = "C:/Windows/Fonts/malgun.ttf" 
# f_path = "/Libary/Fonts/AppliGothic.ttf"
font_name = font_manager.FontProperties(fname = f_path).get_name() 
rc('font', family=font_name)
```

## 성별별 시각화

### 1. 

```python
df.index = df['분류']
```

```python
df[df['대분류']=='성별']
```

```python
df.loc[df['대분류']=='성별', '운동을 할 충분한 시간이 없어서'].plot.pie(autopct="%1.0f")
```

### 2. 선생님

```python
s_df = df[df['대분류']=='성별'].copy()
s_df
```

```python
s_df.drop(columns='대분류', inplace=True)
s_df
```

```python
s_df
```

```python
(f, ax) = plt.subplots(1, 3, figsize=(16, 8)) # 그래프 하나를 1행 3열로 쪼갠다, 가로 16, 세로 8

 # explode 간격, ax 위치, autopct 퍼센트 보여주기
s_df['운동을 할 충분한 시간이 없어서'].plot.pie(explode=[0, 0.02], ax=ax[0], autopct="%1.1f")
s_df['함께 운동을 할 사람이 없어서'].plot.pie(explode=[0, 0.02], ax=ax[1], autopct="%1.0f")
s_df['운동을 싫어해서'].plot.pie(explode=[0, 0.02], ax=ax[2], autopct="%1.0f")

ax[0].set_ylabel('')

plt.show()
```

## 연령대별 시각화

```python
a_df = df[df['대분류']=='연령별'].copy()
a_df
```

```python
a_df.drop(columns='대분류', inplace=True)
```

```python
a_df.drop(columns='분류', inplace=True)
```

```python
a_df
```

```python
(f, ax) = plt.subplots(1, 4, figsize=(16, 8)) # 그래프 하나를 1행 3열로 쪼갠다, 가로 16, 세로 8

 # explode 간격, ax 위치, autopct 퍼센트 보여주기
explodes = [0,0,0,0,0.02,0.09]
a_df['운동을 할 충분한 시간이 없어서'].plot.pie(explode=explodes, ax=ax[0], autopct="%1.1f")
a_df['함께 운동을 할 사람이 없어서'].plot.pie(explode=explodes, ax=ax[1], autopct="%1.0f")
a_df['운동을 할 만한 장소가 없어서'].plot.pie(explode=explodes, ax=ax[2], autopct="%1.0f")
a_df['운동을 싫어해서'].plot.pie(explode=explodes, ax=ax[3], autopct="%1.0f")

ax[0].set_title('운동을 할 충분한 시간이 없어서')
ax[1].set_title('함께 운동을 할 사람이 없어서')
ax[2].set_title('운동을 할 만한 장소가 없어서')
ax[3].set_title('운동을 싫어해서')


ax[0].set_ylabel('')
ax[1].set_ylabel('')
ax[2].set_ylabel('')
ax[3].set_ylabel('')

plt.show()
```
