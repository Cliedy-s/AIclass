```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
```

## 랭킹 검색

```python
headers = {
    'User-Agent' : UserAgent().ie
}

res = requests.get("https://www.melon.com/chart/", headers=headers)
```

> td:nth-child(6) : td의 6번째 자식<br>
td:nth-of-type(n) : n번째 td인 자식 

```python
soup = BeautifulSoup(res.text, 'html.parser')

# html에 class를 여러개 주고싶으면 class="class1 class2 class3 class4"
for n in range(1, 101):
    song = soup.select(f'#frm > div.service_list_song > table > tbody > tr:nth-of-type({n}) > td:nth-of-type(6) a') # 35번째 찾기
    print(n,'위',song[0].string, ' :: ',song[1].string , '\n')
```

# 웹스크래핑 10 : 다나와 크로울링

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.common.by import By
import time
import os
import urllib.request as req
import xlsxwriter # 엑셀 저장
```

## 켜기

```python
browser= webdriver.Chrome('./chromedriver_win32/chromedriver.exe')
browser.implicitly_wait(5)
```

## 다나와 켜기

```python
# browser.set_window_size(1920, 1280)
browser.get('http://prod.danawa.com/list/?cate=112758&15main_11_02')
```

## 조작

### 해당 id를 갖는 버튼이 화면에 보일 때까지 기다려라

```python
# 제조사 더보기 버튼 클릭
# 최대 3초 기다린다.
# copy => xpath
WebDriverWait(browser, 3).until(ec.presence_of_element_located((By.XPATH, '//*[@id="dlMaker_simple"]/dd/div[2]/button[1]'))).click()
# 애플 버튼 클릭
WebDriverWait(browser, 3).until(ec.presence_of_element_located((By.XPATH, '//*[@id="selectMaker_simple_priceCompare_A"]/li[15]'))).click()
# time.sleep(3)
# print(browser.page_source)
```

```python
soup = BeautifulSoup(browser.page_source, 'html.parser')
prodlist = soup.select('div.main_prodlist.main_prodlist_list > ul > li.prod_item.prod_layer')

# 저장할 디렉토리 생성
savePath = 'C:/Users/GD6/imagedownPC/'
if not (os.path.isdir(savePath)):
    os.makedirs(os.path.join(savePath))


 # 문자열 trim == strip
 # text <=> string  : text 는 태그 빼고, string은 해당 태그내부의 문자열
item = prodlist[0]
#User-Agent
opener = req.URLopener()
opener.addheader('User-Agent','whatever')

# 엑셀 처리
workbook = xlsxwriter.Workbook('./data/danawa_crawling_result.xlsx')
worksheet = workbook.add_worksheet()


for n, item in enumerate(prodlist):
    # if not n.find('div',class='adcaster'):
    prodname = item.select('p.prod_name > a')[0].text.strip()
    prodprice = item.select('p.price_sect > a')[0].text.strip()
    prodlink = item.select('a.thumb_link')[0].get('href').strip() # 속성 가져오기
    prodcode = prodlink.split('pcode=')[1][:7]
    prodimg = item.select('a.thumb_link > img')[0].get('data-original').split('?')[0] # 속성 가져오기
    print(prodcode, '::' ,prodname, prodprice,'\n', prodimg)
    
    # 엑셀 파일로 저장 # write(위치)
    worksheet.write(f'A{n+1}', prodcode)
    worksheet.write(f'B{n+1}', prodname)
    worksheet.write(f'C{n+1}', prodprice)
    
    # download retreive=> 오류 => User-Agent설정
    FileName = os.path.join(savePath +prodcode+'.jpg')
    #req.urlretrieve(prodimg, os.path.join(savePath +prodcode+'.jpg'))
    opener.retrieve(prodimg, FileName)
    
workbook.close()
```

```python
browser.quit()
```

# 웹스크래핑 10 : 다나와 크로울링

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.common.by import By
import time
import os
import urllib.request as req
import xlsxwriter # 엑셀 저장
```

## 켜기

```python
browser= webdriver.Chrome('./chromedriver_win32/chromedriver.exe')
browser.implicitly_wait(5)
```

## 다나와 켜기

```python
# browser.set_window_size(1920, 1280)
browser.get('http://prod.danawa.com/list/?cate=112758&15main_11_02')
```

## 조작

### 해당 id를 갖는 버튼이 화면에 보일 때까지 기다려라

```python
# 제조사 더보기 버튼 클릭
# 최대 3초 기다린다.
# copy => xpath
WebDriverWait(browser, 3).until(ec.presence_of_element_located((By.XPATH, '//*[@id="dlMaker_simple"]/dd/div[2]/button[1]'))).click()
# 애플 버튼 클릭
WebDriverWait(browser, 3).until(ec.presence_of_element_located((By.XPATH, '//*[@id="selectMaker_simple_priceCompare_A"]/li[15]'))).click()
# time.sleep(3)
# print(browser.page_source)
```

### 한번만 실행해도 됨

```python
# 저장할 디렉토리 생성
savePath = 'C:/Users/GD6/imagedownPC/'
if not (os.path.isdir(savePath)):
    os.makedirs(os.path.join(savePath))


 # 문자열 trim == strip
 # text <=> string  : text 는 태그 빼고, string은 해당 태그내부의 문자열
item = prodlist[0]
#User-Agent
opener = req.URLopener()
opener.addheader('User-Agent','whatever')

# 엑셀 처리
workbook = xlsxwriter.Workbook('./data/danawa_crawling_result.xlsx')
worksheet = workbook.add_worksheet()
```

```python
#현재 페이지 수
cur_page_num = 1
#전체 페이지 수
target_page_num = 7


row_num=1
while cur_page_num <= target_page_num:
    soup = BeautifulSoup(browser.page_source, 'html.parser')
    prodlist = soup.select('div.main_prodlist.main_prodlist_list > ul > li.prod_item.prod_layer')

    for item in prodlist:
        # if not n.find('div',class='adcaster'):
        prodname = item.select('p.prod_name > a')[0].text.strip()
        prodprice = item.select('p.price_sect > a')[0].text.strip()
        prodlink = item.select('a.thumb_link')[0].get('href').strip() # 속성 가져오기
        prodcode = prodlink.split('pcode=')[1][:7]
        prodimg = item.select('a.thumb_link > img')[0].get('data-original').split('?')[0] # 속성 가져오기
        print(prodcode, '::' ,prodname, prodprice,'\n', prodimg)

        # 엑셀 파일로 저장 # write(위치)
        worksheet.write(f'A{row_num}', prodcode)
        worksheet.write(f'B{row_num}', prodname)
        worksheet.write(f'C{row_num}', prodprice)
        row_num +=1

        # download retreive=> 오류 => User-Agent설정
        FileName = os.path.join(savePath +prodcode+'.jpg')
        #req.urlretrieve(prodimg, os.path.join(savePath +prodcode+'.jpg'))
        opener.retrieve(prodimg, FileName)

    print()
    cur_page_num += 1
    if(cur_page_num > target_page_num):
        print('크로울링 끝')
        break;
    # 페이지 이동버튼 클릭
    WebDriverWait(browser, 3).until(ec.presence_of_element_located((By.CSS_SELECTOR, f'div.number_wrap > a:nth-child({cur_page_num})'))).click()
    
    del soup
    time.sleep(5)

workbook.close()
```

```python
browser.quit()
```
