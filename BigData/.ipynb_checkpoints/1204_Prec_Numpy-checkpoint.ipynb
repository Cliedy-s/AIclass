### 퀴즈<02>
./data/movielens-1m/ratings.dat 파일을 읽고 데이터 분석하기


- 전체 평균 평점 계산하기
- 각 사용자 별 평균 평점 계산하기
- ./myfile/mean_rating_by_user.csv 파일에 저장하기

```python
import numpy as np
data = np.loadtxt("data/movielens-1m/ratings.dat", delimiter='::', dtype=np.int64)
# 데이터 받아오기
```

```python
# 전체 평균 평점
round(data[:,2].mean(),2)
```

```python
# 각 사용자 별 레이팅
user_ids = np.unique(data[:, 0]) # 첫번째 열의 distinct

user_rate = []
for userid in user_ids:
    user_rate.append((userid, data[data[:,0] == userid][:,2]))

user_rate
```

```python
# 레이팅 평균
user_rateAvgs = []
for (x, y) in user_rate:
    user_rateAvgs.append([x, round(y.mean(),2)])
    
user_rateAvgs
```

```python
np.savetxt("myfile/mean_rating_by_user.csv", user_rateAvgs, fmt='%-10d %.3f', delimiter=',')
# 데이터 형식에 맞춰 저장
```
