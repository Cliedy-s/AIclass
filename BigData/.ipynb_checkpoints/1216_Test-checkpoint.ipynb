## 상가업소정보 2019년 6월

```python
shop_df = pd.read_csv('./data/commercial_store/소상공인시장진흥공단_상가업소정보_201906_01.csv')
shop_df.head()
```

```python
shop_df.columns
view_columns =['상호명', '지점명', '상권업종대분류명',
               '상권업종중분류명', '상권업종소분류명','시도명',
               '시군구명',  '행정동명', '법정동명','지번주소',
               '경도', '위도']
shop_df[view_columns].head()
```

```python
shop_df['상권업종중분류명'].unique()
```

```python
df_seoul = shop_df.loc[shop_df['지번주소'].str.startswith('서울')]
df_cafe = df_seoul[df_seoul['상권업종중분류명']=='제과제빵떡케익']
df_bread = df_cafe[df_cafe['상호명'].str.contains('뚜레쥬르|파리바게트|Tous Les Jours|tous les jours|TOUS LES JOURS|파리바게뜨|paris baguette|PARIS BAGUETTE')]
df_bread[view_columns].head()
```

## 뚜레쥬르와 파리바게트 총 업소 수

```python
df_bread['상호명'].value_counts().sum()
```

### 뚜레쥬르 업소 수

```python
df_bread.loc[df_bread['상호명'].str.contains('뚜레쥬르|Tous Les Jours|tous les jours|TOUS LES JOURS'), '상호명'].value_counts().sum()
```

### 파리바게트 업소 수

```python
df_bread.loc[df_bread['상호명'].str.contains('파리바게트|파리바게뜨|paris baguette|PARIS BAGUETTE'), '상호명'].value_counts().sum()
```

## 브랜드명 생성

```python
df_bread[df_bread['상호명'].str.contains('뚜레쥬르|Tous Les Jours|tous les jours|TOUS LES JOURS')]['브랜드명'] = '뚜레쥬르'
df_bread['브랜드명'] = df_bread['브랜드명'].fillna('파리바게트')
```

```python
df_bread.head()
```

```python
df_bread['브랜드명'].value_counts()
```

```python
df_bread_brand=pd.DataFrame(df_bread.groupby(['시군구명','브랜드명'])['상호명'].count()).reset_index()
df_bread_brand.head()
```

```python
df_bread_brand.columns = ['구','브랜드명','매장수']
df_bread_brand.head()
```

## 구별 매장 수 비교

```python
plt.figure(figsize=(24,6))
sns.barplot(data=df_bread_brand, x='구', y='매장수', hue='브랜드명')
plt.show()
```

## 지도

```python
plt.figure(figsize=(16,12))
sns.scatterplot(data=df_bread,x='경도', y='위도', hue='브랜드명')
```

```python
mym = folium.Map(location=[df_bread['위도'].mean(), df_bread['경도'].mean()], zoom_start=11, tiles='Stamen Toner')
for n in df_bread.index:
    if df_bread.loc[n, '브랜드명']=='파리바게트' :
        icon_color='orange'
    else:
        icon_color='blue'
    folium.CircleMarker(
        location=[df_bread.loc[n, '위도']
        , df_bread.loc[n, '경도']]
        , radius=2
        , popup=df_bread.loc[n, '상호명']
        , color=icon_color
        , fill=True
        , fill_color=icon_color
    ).add_to(mym)
    
mym
```

## 가장 매장이 많은 구의 매장 구성

### 가장 매장이 많은 구 top3

```python
df_bread_brand_total=pd.DataFrame(df_bread.groupby('시군구명')['상호명'].count()).reset_index()
df_bread_brand_total.head()
```

```python
top3 = df_bread_brand.sort_values(by='상호명', ascending=False).head(3)['시군구명']
top3
```

```python
pivotbread = df_bread_brand.pivot('브랜드명','시군구명','상호명')
pivotbread
```

```python
(f, ax) = plt.subplots(1, 3, figsize=(16, 8))

pivotbread[top3.iloc[0]].plot.pie(explode=[0, 0.05], ax=ax[0], autopct="%1.0f")
pivotbread[top3.iloc[1]].plot.pie(explode=[0, 0.05], ax=ax[1], autopct="%1.0f")
pivotbread[top3.iloc[2]].plot.pie(explode=[0, 0.05], ax=ax[2], autopct="%1.0f")

ax[0].set_title(top3.iloc[0])
ax[0].set_ylabel('')
ax[1].set_title(top3.iloc[1])
ax[1].set_ylabel('')
ax[2].set_title(top3.iloc[2])
ax[2].set_ylabel('')

plt.show()
```

# 매장이 적은 구의 매장 구성

### 가장 매장이 적은 구 top3

```python
df_bread_brand_total=pd.DataFrame(df_bread.groupby('시군구명')['상호명'].count()).reset_index()
df_bread_brand_total.head()
```

```python
top3 = df_bread_brand.sort_values(by='상호명', ascending=True).head(3)['시군구명']
top3
```

```python
(f, ax) = plt.subplots(1, 3, figsize=(16, 8))

pivotbread[top3.iloc[0]].plot.pie(explode=[0, 0.05], ax=ax[0], autopct="%1.0f")
pivotbread[top3.iloc[1]].plot.pie(explode=[0, 0.05], ax=ax[1], autopct="%1.0f")
pivotbread[top3.iloc[2]].plot.pie(explode=[0, 0.05], ax=ax[2], autopct="%1.0f")

ax[0].set_title(top3.iloc[0])
ax[0].set_ylabel('')
ax[1].set_title(top3.iloc[1])
ax[1].set_ylabel('')
ax[2].set_title(top3.iloc[2])
ax[2].set_ylabel('')

plt.show()
```

# 총평

뚜레쥬르와 파리바게트의 매장 수는 매우 차이가 있었고 뚜레쥬르의 매장 수가 파리바게트의 매장 수를 넘는 구는 존재하지 않았다.
지도를 보았을 때도 파리바게트가 중심에 몰리기도 몰리고 넓게도 퍼져있었다.
그에 반해 뚜레쥬르는 적당히 퍼져있다.
